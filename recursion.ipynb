```python
#find factorial of 5
def factorial(n):
    if n==1 or n==0:
        return 1
    return n*factorial(n-1)
factorial(5)    
```

```python
#find sum of n natural numbers using recursion
def natural_number(n):
    if n==1:
        return 1
    return n*(n+1)/2
natural_number(4)

```

```python
#binary number for decimal number
def binary(n):
    if n==0:
        return "0"
    if n==1:
        return "1"
    return binary(n//2)+str(n%2) 
binary(6)
```

```python
square = lambda x : x*x
print(square(6))
```

```python
#sum of 2 numbers using lambda
add=lambda x,y: x+y
print(add(3,4))
```

```python
# find square of all list of integer using lambda and map()
arr=[1,2,3,4,5]
res=list(map(lambda x: x*2, arr))
print(res)
```

```python
#find the even number from list using lambda  and filter
arr=[1,2,3,4,5]
res=list(filter(lambda x: x%2==0, arr))
print(res)
```

```python
# find the number is even or odd using lambda
n=int(input())
even_or_odd=lambda x:f"{x}odd" if x%2 else f"{x}-even"
print(even_or_odd(n))
```
