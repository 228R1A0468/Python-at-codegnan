```python
#sorted(iterable, reverse=False, key=None):
#sort the fruits based on the length of fruit name
fruit=["Apple","banana","kivi","Mango","Water melon"]
res=sorted(fruit)
print(res)
print(sorted(fruit,reverse=True))
print(sorted(fruit,reverse=True,key=len))
print(sorted(fruit,reverse=True,key=min))
```

```python
#map(function,iterable)

#find the square of all list of integers using map function
def square(x):
    return x**2
    
arr=[1,2,3,4,5]
res=list(map(square,arr))
print(res)
```

```python
#filter(function, iterable)
#find the even number from list ogf numbers
def check_even(x):
    if x%2==0:
        return x
lst=[1,2,3,4,5,6]
res=list(filter(check_even,lst))
print(res)
```

```python
def check_even(x):
    if x%2==0:
        return x
lst=[1,2,3,4,5,6]
res=list(map(check_even,lst))
print(res)
```

```python
#find the 3's multiplers from list of integers
#arr=[1,24,52,33,39,15,30]
def check_multiplers(x):
    if x%3==0:
        return x
arr=[1,24,52,33,39,15,30]  
res=list(filter(check_multiplers,arr))
print(res)
```

```python
# enumerate(iterable)
fruits=["Apple","banana","kivi","Mango","Water melon"]
print(enumerate(fruits))
print(list(enumerate(fruits)))
for ind, val in enumerate(fruits):
    print(f"{ind}-{val} and length is {len(val)}")
```

```python
percentage=[30,40,50,20]
for num in percentage:
    if num<35:
        print("All students not passed")
        break
else:
    print("All students pass")
```

```python
percentage=[30,40,50,20]
lst=[]
for num in percentage:
    if num<35:
        lst.append(False)
    else:
        lst.append(True)
print(lst)
if all(lst):
    print("All students are passed")
else:
    print("All students not pass")
```
