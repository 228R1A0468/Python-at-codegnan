```python
#with parameter and return
#sum of 2 numbers
def add(x:int,y:int)->int:
    c=x+y
    return c
res=add(3,5)
print(res)
    
```

```python
#with return and without parameters
def add():
    x,y=20,39
    c=x+y
    return c
res=add()
print(res)
print(add())
print(add())
```

```python
# without return and with parameters
#when fuction not returns any value or fuction doesnt contains return statement in this case,functions by default it returnd "None" as result to fuction call block
def add(x:int,y:int)->int:
    c=x+y
    print(f" this id from fuction defination block: {c}")
    
res=add(3,5)
print(f" this id from fuction defination block: {res}")
    
```

```python
#find the wheather the number is prime number or not using function
#logic 2, n//2+1
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True

# Print prime numbers from 1 to 100
print("Prime numbers from 1 to 100 are:")

for num in range(1, 101):
    if is_prime(num):
        print(num, end=" ")
    
```
