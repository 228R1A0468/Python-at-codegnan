```python
# positional arguments
def sub(x:int,y:int):
    c=x-y
    return c
a,b=10,20
res=sub(a,b)
print(res)
print(sub(20,10))
```

```python
 #keyword arguments
def sub(x:int,y:int):
    c=x-y
    return c
a,b=10,20

print(sub(x=a,y=b))
print(sub(y=b,x=a))
```

```python
def add (x:int,y:int,s:int,p:int):
    d=x+y
    f=x+y+s
    v=x+y+s+p
    return d,f,v
a,b,c,o=10,20,30,40
res=add(a,b,c,o)
print(res)
    
```

```python
# default arguments
def add(a,b,c=0,d=0):
    res=a+b+c+d
    return res
print(add(10,20))    
print(add(10,20,5))    
print(add(10,20,2,1))    
```

```python
def add(a,b,c=1,d=1):
    res=a*b*c*d
    return res
print(add(10,20))    
print(add(10,20,5))    
print(add(10,20,2,1))    
```
