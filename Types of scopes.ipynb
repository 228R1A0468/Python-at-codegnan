```python
#local scope
def fun(x):
    x=x+1  # is local scope
    return x
a=10
res=fun(a)
print(a)
```

```python
# enclosing scope
def outer(x):
    y=10 #enclose scope
    def inner():
        a=10  #local scope
        z=y+a
        print(f"This local scope:{z}")
    inner()
    y=y+x
    print(f"This is enclosing scope:{y}")
n=5
outer(n)
    
```

```python
# enclosing scope
def outer(x):
    x=10 #enclose scope
    def inner():
        x=20  #local scope
        z=x+x
        print(f"This local scope:{z}")
    inner()
    y=x+x
    print(f"This is enclosing scope:{y}")
n=5
outer(n)
    
```

```python
# global scope
x=10 # global scope
def outer():
    y=20 3 enclosing scope
    def inner():
        z=30 # local scope
        res=z+y+x
        print(f"This is local scope: {res}")
    res=y+x
    inner()
    print(f"This is enclosing scope:{res}")
outer()

      
```
