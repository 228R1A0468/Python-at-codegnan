```python
# finding even or not using contional stm
n1,n2,n3=map(int,input().split())
if n1%2 :
    print(f"{n1} is odd")
else:
    print(f"{n1} is even")
if n2%2 :
    print(f"{n2} is odd")
else:
    print(f"{n2} is even") 
if n3%2 :
    print(f"{n3} is odd")
else:
    print(f"{n3} is even")    
```

```python
# finding even or odd using loop
for val in [11,12,13]:
    if val%2==0:
        print(f"{val} is even")
    else:
        print(f"{val} is odd")
else:
    print("program completed")
```
