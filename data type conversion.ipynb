```python
#reading float value from user
#float to int, string,bool
val=float(input())
print(type(val))
print(int(val))
print(str(val)[:2])
print(bool(val))
```

```python
#reading string from user
#string to int,float,bool,(when input is int)
s=input()
print(type(s),s)
print(int(s))
print(float(s))
print(bool(s))
```

```python
#when input is string
s=input()
print(type(s),s)
print(list(s)[4:-3:1])
print(tuple(s))
print(set(s))
```

```python
#reading list from user
#list to tuple,set,string

n=list(input())
print(type(n),n)
print(tuple(n)[1:-6:1])
print(set(n))
print(str(n))
print(list(map(int,input().split())))  #when we want convert list to int
print(list(map(str,input().split())))  #vice verca
```

```python
# readlist of strings
l=list(map(str,input().split()))
print(l)
```

```python
# reading tuple of integers
t=tuple(map(int,input().split()))
print(list(t))
print(set(t))
print(str(t))
        
```

```python
# reading set of integer

t=set(map(int,input().split()))
print(list(t))
print(tuple(t))
print(str(t))
        
```

```python
# eval() it evalutes the sting type expression
print("3+5")
print(3+5)
print(eval("3+5"))
```
