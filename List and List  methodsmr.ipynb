```python
# list is mutable
list=[1,2,3,4,8.9,"akshitha"]
list[2]=100
print(list)
```

```python
# indexing and slicing
list=[1,2,23,4.56,"akshitha"]
print(list[-1])
print(list[-3])
print(list[:100])
print(list[-1:5:-1])
print(list[-10:-1:-1])
print(list[::-2])
print(list[::-1])
```

```python
# element removing methods
#remove
lst=[1,2,3,4,6]
lst.remove(2)
print(lst)
#pop
val =lst.pop()
print(val,lst)
val2=lst.pop(1)
print(val,lst)
#del
del lst([0])
print(lst)
lst.clear()
print(lst)

```

```python
#list methods add methods
#append
lst=[1,2,3,4]
(lst.append(50))
(lst.append("code"))
lst.append([10,20,30])
print(lst)
print(len(lst))

#expand
lst.extend([10,29,30])
print(lst)
print(len(lst))

#insert
lst.insert(1,10000)
print(lst)
lst.insert(0,10000)
print(lst)
```
