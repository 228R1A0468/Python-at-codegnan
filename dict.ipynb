```python
#dict
details={"name":"ravi","age":20,"course":"python"}
print(details)
```

```python
#dict cant allow list as a key
details={"name":"ravi","age":20,"course":"python","no":[1,2,3]}
print(details)
```

```python
#key should be different it doest allow duplicates but values can
#if dict contains multiple item with key, in that case it updates the first item key value with last defined item value
details={"name":"ravi","age":20,"course":"python","no":[1,2,3],"age":50}
print(details)
```

```python
# access the values by using the keys
details={"name":"ravi","age":20,"course":"python","no":[1,2,3],"age":50}
print(details["course"])
print(details["age"])
#updating thr dict value
details["age"]+=1
details["course"]="java" # it updates the value for course key
print(details["age"])
print(details["course"])
details["location"]="india" #it adds new item to dict
print(details)
```

```python
# define empty dit
d1={}
d2=dict()
print(type(d1),type(d2))
```

```python
# dict methods
# get method
details={"name":"ravi","age":20,"course":"python","no":[1,2,3],"age":50}
print(details.get("name"))
print(details.get("location"))
print(details.get("age",18))
print(details.get("loction", "location key not exists"))
```

```python
#update method
details={"name":"ravi","age":20,"course":"python"}
d={"marks":[85,90,86], "age":50}
details.update(d)
print(details)

```

```python
#pop method
details={"name":"ravi","age":20,"course":"python","no":[1,2,3],"age":50}

print(details.pop("age"))
print(details)
print(details.popitem())
print(details)
```

```python
details={"name":"ravi","age":20,"course":"python","no":[1,2,3],"age":50}
keys=details.keys()
print(type(keys))
print(details.keys())         
print(details.values())
print(details.items())
print(list(keys))
print(tuple(keys))

```
